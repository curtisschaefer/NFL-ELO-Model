```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from datetime import date
from datetime import timedelta
import statsmodels.formula.api as smf
from scipy.stats import gaussian_kde
from scipy.stats import norm
import statistics
from scipy import stats
from scipy.optimize import minimize
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
from matplotlib.patches import Rectangle
import os
import re
```

```python


# Step 1: READ IN play by play records, convert to mlb_outcomes, don't run this unless you need to
# instead run the cell below 

directory_path = '../Baseball_Data/BR/X/PBP/'

# Step 2: Get a list of all CSV files in the directory that match the pattern
csv_files = [file for file in os.listdir(directory_path) if file.endswith('.csv')]

csv_files = [s for s in csv_files if int(s[3:7]) > 2011]

# Initialize an empty list to store DataFrames
dfs = []

# Step 3: Read each CSV file, add a 'code' column, and append it to the list of DataFrames
count = 0
for csv_file in csv_files:
    code = csv_file.split('.')[0]  # Extract the filename without the extension
    file_path = os.path.join(directory_path, csv_file)
    
    try:
        df = pd.read_csv(file_path)
        df['code'] = code  # Add a 'code' column with the extracted filename
        df['vis'] = df.iloc[0]['@Bat']
        df['home'] = df[df['Inn'].str.startswith('b', na=False)].iloc[0]['@Bat']
        df[['batscore', 'fieldscore']] = df['Score'].str.split('-', expand=True)
        df['batscore'] = df['batscore'].astype(int)
        df['fieldscore'] = df['fieldscore'].astype(int)
        df['batscore'] = df.batscore + df['R/O'].str.count('R')
        df['home_score'] = np.where(df['Inn'].str.startswith('b', na=False),df['batscore'],df['fieldscore'])
        df['vis_score'] = np.where(df['Inn'].str.startswith('t', na=False),df['batscore'],df['fieldscore'])
        df['margin'] = np.abs(df['home_score'] - df['vis_score'])
        df['month'] = df['code'].str[7:9].astype(int)  # Extract '04' as integer
        df['day'] = df['code'].str[9:11].astype(int)   # Extract '05' as integer
        df['second'] = df['code'].str[11:12].astype(int) # Extract trailing '0' as string
        df['year'] = df['code'].str.slice(3, 7)
        dfs.append(df)
        count += 1  # Increment the count of successfully processed files
    except Exception as e:
        print(f"Error processing {csv_file}: {e}")

# Optional: Combine all DataFrames into one if needed
if dfs:
    cdf = pd.concat(dfs, ignore_index=True)
    print(f"Successfully combined {count} files into a single DataFrame.")
else:
    print("No files were processed.")

cdf = cdf.loc[cdf.groupby('code')['playnum'].idxmax()]
cdf['winner'] = np.where(cdf.home_score > cdf.vis_score,cdf['home'],cdf['vis'])
cdf = cdf[['vis','home','year','month','day','second','winner','margin']]
cdf = cdf.sort_values(by=['year','month', 'day', 'second'], ascending=[True,True, True, True])

elo_dict = {key: 1500 for key in cdf['home'].unique().tolist()}

cdf.to_csv('./mlb_outcomes.csv', index=False)
```

```python


# Define the function to return the log likelihood
def compute_log_likelihood(k, cdf):
    """
    Compute the total negative log-likelihood for a given K-value in an Elo model.
    
    Parameters:
        k (float): The K-value to use for Elo rating updates (passed by minimize).
        cdf (DataFrame): A DataFrame containing game data with columns
                         ['vis', 'home', 'winner'].

    Returns:
        float: The negative sum of log-likelihood for all games in the DataFrame.
    """
    # Initialize Elo dictionary with a default rating of 1500
    elo_dict = {team: 1500 for team in pd.concat([cdf['vis'], cdf['home']]).unique()}
    
    # Initialize the outcome log-likelihood
    log_likelihood = 0
    
    # Process each game in the DataFrame
    for _, row in cdf.iterrows():
        vis = row['vis']       # Visiting team
        home = row['home']     # Home team
        winner = row['winner'] # Winner of the game

        # Get the current Elo ratings
        vis_elo = elo_dict[vis]
        home_elo = elo_dict[home]

        # Calculate win probabilities for both teams
        vis_prob = 1 / (1 + 10 ** ((home_elo - vis_elo) / 400))
        home_prob = 1 - vis_prob

        # Determine the likelihood of the realized outcome
        if winner == vis:
            likelihood = vis_prob
        elif winner == home:
            likelihood = home_prob
        else:
            raise ValueError("Unexpected winner value in row.")

        # Add log-likelihood for this game
        log_likelihood += np.log(likelihood)

        # Determine the loser
        loser = vis if winner != vis else home

        # Update Elo ratings
        elo_dict[winner] += k * (1 - (vis_prob if winner == vis else home_prob))
        elo_dict[loser] += k * (0 - (home_prob if loser == home else vis_prob))
    
    return -log_likelihood  # Negative for minimization


# check the log likelihood for a value

print(compute_log_likelihood(6.25,cdf))


# Define the objective function for optimization
# It now takes k and hfa as parameters
def objective(params, cdf):
    k = params  # Extract k, and hfa from parameters
    return compute_log_likelihood(k,  cdf)

# Run optimization over k and hfa
result = minimize(
    lambda params: objective(params, cdf),
    x0=[30],  # Initial guesses for k, t, and hfa
    bounds=[(0, None)],  # Bounds: k >= 0,  hfa >= 0
    method='L-BFGS-B'
)

# Extract MLE for k, t, and hfa
optimal_k = result.x
print(f"Optimal K-value: {optimal_k}")


# Compute confidence intervals using the Hessian
hessian_inv = result.hess_inv.todense()  # Inverse Hessian
std_error = np.sqrt(np.diag(hessian_inv))  # Standard errors for k and hfa

# 95% Confidence Intervals
z_value = 1.96
lower_bound_k = optimal_k - z_value * std_error[0]
upper_bound_k = optimal_k + z_value * std_error[0]


print(f"95% Confidence Interval for K: ({lower_bound_k}, {upper_bound_k})")

```

```python
### compute the linear model 


# Define the function to return the log likelihood
def compute_MSE(k_one,sigma_one, cdf):
    
    # Initialize Elo dictionary with a default rating of 1500
    elo_dict = {team: 1500 for team in pd.concat([cdf['vis'], cdf['home']]).unique()}
    
    # Initialize the outcome log-likelihood
    mse = 0
    
    # Process each game in the DataFrame
    for _, row in cdf.iterrows():
        vis = row['vis']       # Visiting team
        home = row['home']     # Home team
        margin = row['margin'] # margin of the game
        winner = row['winner'] # Winner of the game
        # Determine the loser
        loser = vis if winner != vis else home

        

        # Calculate predicted margin 
        mhat_winner = (elo_dict[winner] - elo_dict[loser])/sigma_one

        # Add to mse for this game
        mse += (margin - mhat_winner)**2 


        # Update Elo ratings
        elo_dict[winner] += k_one * (margin - mhat_winner)
        elo_dict[loser] += k_one * (-1*(margin - mhat_winner))
        

    #print(dict(sorted(elo_dict.items(), key=lambda item: item[1])))

    
    return mse  


# check the mse for a value

print(compute_MSE(4.66,551,cdf))



# Define the objective function for optimization
# It now takes k and hfa as parameters
def objective(params, cdf):
    k, sig = params  # Extract k, and hfa from parameters
    return compute_MSE(k,  sig, cdf)

# Run optimization over k and hfa
result = minimize(
    lambda params: objective(params, cdf),
    x0=[30, 500],  # Initial guesses for k, t, and hfa
    bounds=[(0, None),  (0, None)],  # Bounds: k >= 0,  hfa >= 0
    method='L-BFGS-B'
)

# Extract MLE for k, t, and hfa
optimal_k,  optimal_sig = result.x
print(f"Optimal K-value: {optimal_k}")
print(f"Optimal sig: {optimal_sig}")


# Compute confidence intervals using the Hessian
hessian_inv = result.hess_inv.todense()  # Inverse Hessian
std_error = np.sqrt(np.diag(hessian_inv))  # Standard errors for k and hfa

# 95% Confidence Intervals
z_value = 1.96
lower_bound_k = optimal_k - z_value * std_error[0]
upper_bound_k = optimal_k + z_value * std_error[0]
lower_bound_sig = optimal_sig - z_value * std_error[1]
upper_bound_sig = optimal_sig + z_value * std_error[1]

print(f"95% Confidence Interval for K: ({lower_bound_k}, {upper_bound_k})")
print(f"95% Confidence Interval for sig: ({lower_bound_sig}, {upper_bound_sig})")

```

```python
### compute the joint additive model 


# Define the function to return the log likelihood
def compute_JA_loss(k_one,k_two, cdf):

    sigma_one = 551 # derived from linear model 
    
    # Initialize Elo dictionary with a default rating of 1500
    elo_dict = {team: 1500 for team in pd.concat([cdf['vis'], cdf['home']]).unique()}
    
    # Initialize the outcome log-likelihood
    mse = 0
    log_likelihood = 0
    
    # Process each game in the DataFrame
    for _, row in cdf.iterrows():
        vis = row['vis']       # Visiting team
        home = row['home']     # Home team
        margin = row['margin'] # margin of the game
        winner = row['winner'] # Winner of the game
        # Determine the loser
        loser = vis if winner != vis else home
        
        # Get the current Elo ratings
        vis_elo = elo_dict[vis]
        home_elo = elo_dict[home]

        # Calculate predicted margin 
        mhat_winner = (elo_dict[winner] - elo_dict[loser])/sigma_one

        # Add to mse for this game
        mse += (margin - mhat_winner)**2 

        # Calculate win probabilities for both teams
        vis_prob = 1 / (1 + 10 ** ((home_elo - vis_elo) / 400))
        home_prob = 1 - vis_prob

        # Determine the likelihood of the realized outcome
        if winner == vis:
            likelihood = vis_prob
        elif winner == home:
            likelihood = home_prob
        else:
            raise ValueError("Unexpected winner value in row.")

        # Add log-likelihood for this game
        log_likelihood += np.log(likelihood)

        # Update Elo ratings
        elo_dict[winner] += k_one * (margin - mhat_winner)
        elo_dict[loser] += k_one * (-1*(margin - mhat_winner))
        elo_dict[winner] += k_two * (1 - (vis_prob if winner == vis else home_prob))
        elo_dict[loser] += k_two * (0 - (home_prob if loser == home else vis_prob))
        
    stdev = cdf['margin'].std()
    loss_function = np.sqrt(mse)  #-   log_likelihood

    #print(dict(sorted(elo_dict.items(), key=lambda item: item[1])))

    
    return loss_function  


# check the mse for a value

print(compute_JA_loss(4.66,6.25,cdf))
print(compute_JA_loss(4.66,4.66,cdf))
print(compute_JA_loss(5.66,4.66,cdf))
```
